## 3.7 Model Performance Definition

To accurately assess the effectiveness of the models, multiple performance metrics are employed:

- **Accuracy**: Measures the proportion of correctly classified instances over the total number of samples. It provides a general sense of model performance but may be misleading in imbalanced datasets.
  
  $$
  \text{Accuracy} = \frac{\text{True Positives} + \text{True Negatives}}{\text{Total Samples}}
  $$
  
- **Confusion Matrix**: A detailed breakdown of correct and incorrect classifications across each class, highlighting the types of errors made by the model.

  $$
  \begin{array}{cc}
  \text{True Positives (TP)} & \text{False Positives (FP)} \\
  \text{False Negatives (FN)} & \text{True Negatives (TN)} \\
  \end{array}
  $$
  
- **Sensitivity (Recall)**: Indicates the model's ability to correctly identify positive instances (true stories).

  $$
  \text{Sensitivity} = \frac{\text{TP}}{\text{TP} + \text{FN}}
  $$
  
- **Specificity**: Measures the model's ability to correctly identify negative instances (false stories).

  $$
  \text{Specificity} = \frac{\text{TN}}{\text{TN} + \text{FP}}
  $$
  
- **Precision**: Reflects the proportion of positive identifications that were actually correct.

  $$
  \text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}
  $$
  
- **F1 Score**: The harmonic mean of precision and recall, providing a balance between the two.

  $$
  \text{F1 Score} = 2 \times \frac{\text{Precision} \times \text{Sensitivity}}{\text{Precision} + \text{Sensitivity}}
  $$
  
- **Receiver Operating Characteristic (ROC) Curve and Area Under the Curve (AUC)**: Evaluates the model's ability to distinguish between classes across different threshold settings.

  $$
  \text{AUC} = \int_{0}^{1} \text{ROC Curve}
  $$
  
These metrics collectively offer a comprehensive view of the models' strengths and weaknesses, enabling informed decisions on model selection and potential improvements.

## 4.1 Transformation stage

The Transformation Stage converts raw audio recordings and their associated metadata into structured numerical feature vectors essential for machine learning models. This process involves extracting a variety of acoustic and linguistic features.

````python
from abc import ABC, abstractmethod
import numpy as np
from utils.attributes import Audio

class Transformator(ABC):
    """
    Abstract class for all transformators.
    """
    @classmethod
    @abstractmethod
    def transform(cls, dataset: list[Audio]) -> np.ndarray:
        """
        Transform a list of Audio objects into a numpy array.

        Args:
            dataset: list of Audio objects

        Returns:
            np.ndarray: transformed dataset
        """
        pass
````

The `PlainTransformator` function implemented a simple transmator, which includes Mel-Frequency Cepstral Coefficients (MFCCs) to capture the speech spectrum, Chroma features for tonal analysis, Mel Spectrograms to represent time-frequency characteristics, Zero-Crossing Rate (ZCR) for detecting signal noisiness, and Spectral Centroid to assess the brightness of the audio signal. These features selected are commonly used in audio analysis tasks and provide a comprehensive representation of the audio data.

````python
from tqdm import tqdm
import librosa


class PlainTransformator(Transformator):
    """
    Plain transformator that extracts MFCCs, Chroma, Mel Spectrogram, Zero-Crossing Rate, and Spectral Centroid.
    """
    @staticmethod
    def __transform_single(data: Audio) -> np.ndarray:
        """
        Transform a single Audio object into a feature vector.
         
        Args:
            data: Audio object

        Returns:
            np.ndarray: feature vector
        """
        y, sr, language = data.audio, data.sr, data.language

        # Extract MFCCs
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

        # Extract Chroma feature
        chroma = librosa.feature.chroma_stft(y=y, sr=sr)

        # Extract Mel Spectrogram
        mel = librosa.feature.melspectrogram(y=y, sr=sr)

        # Extract Zero-Crossing Rate
        zcr = librosa.feature.zero_crossing_rate(y)

        # Extract Spectral Centroid
        spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)

        # Combine all features into a single feature vector
        feature_vector = np.vstack((
            mfccs,
            chroma,
            mel,
            zcr,
            spectral_centroid
        )).T

        feature_vector = np.hstack((feature_vector, np.full((feature_vector.shape[0], 1), language)))
        return feature_vector

    @classmethod
    def __transform_batch(cls, dataset: list[Audio]) -> np.ndarray:
        """
        Transform a list of Audio objects into a numpy array.
        
        Args:
            dataset: list of Audio objects

        Returns:
            np.ndarray: transformed dataset
        """
        features = []
        
        for data in tqdm(dataset):
            feature = cls.__transform_single(data)
            features.append(feature)
        return np.array(features)
    
    @classmethod
    def transform(cls, dataset):
        return cls.__transform_batch(dataset)
````

`Wav2VecTransformator` utilizes neural network-based feature extraction methods to encapsulate deeper speech patterns. By leveraging a pre-trained deep learning model, Wav2Vec captures intricate temporal and spectral characteristics of the audio recordings, generating rich, high-dimensional feature representations. These advanced features enable the machine learning models to identify subtle nuances and patterns indicative of deception that traditional feature extraction methods might overlook. Consequently, the Wav2VecTransformator enhances the Transformation Stage by providing comprehensive and informative inputs, thereby improving the accuracy and robustness of the subsequent classification tasks.

````python
from pathlib import Path
from tqdm import tqdm, trange


class Wav2VecTransformator(Transformator):
    @staticmethod
    def __process_batch(batch: list[Audio], batch_idx: int, save_file: Path):
        """
        Process a batch of Audio objects and save the features to a file.

        Args:
            batch: list of Audio objects
            batch_idx: index of the batch
            save_file: path to save the features
        """
        import torch, torchaudio
        import numpy as np

        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        
        bundle = torchaudio.pipelines.WAV2VEC2_ASR_BASE_10M
        model = bundle.get_model().to(device)
        model.eval()
        
        features = []
        
        for value in tqdm(batch, desc=f"Batch {batch_idx}"):
            y, sr = value.audio, value.sr

            waveform = torch.from_numpy(y).to(device)
            waveform = waveform.unsqueeze(0)

            if sr != bundle.sample_rate:
                waveform = torchaudio.functional.resample(waveform, sr, bundle.sample_rate)

            with torch.inference_mode():
                feats, _ = model.extract_features(waveform)
                feature_vector = feats[-1][0]
            feature_vector = feature_vector.cpu().numpy()
            
            features.append(feature_vector)
        
        np.save(save_file, np.array(features))
    
    @classmethod
    def transform(cls, dataset) -> np.ndarray:
        """
        Transform a list of Audio objects into a numpy array.

        Args:
            dataset: list of Audio objects

        Returns:
            np.ndarray: transformed dataset
        """
        from threading import Thread

        feature_chunks = []

        save_dir = Path.cwd() / "temp"
        save_dir.mkdir(exist_ok=True)
        
        batch_size = 50
        batch_count = len(dataset) // batch_size
        file_format = "features_batch_{batch_idx}.npy"
        
        for batch_idx in range(batch_count):
            batch = dataset[batch_idx * batch_size: (batch_idx + 1) * batch_size]
            save_file = save_dir / file_format.format(batch_idx=batch_idx)

            # Process the batch in a separate thread, then join the thread
            th = Thread(target=cls.__process_batch, args=(batch, batch_idx, save_file))
            th.start()
            th.join()
        
        for idx in trange(batch_count, desc="Loading features"):
            save_file = save_dir / file_format.format(batch_idx=idx)
            batch_features = np.load(save_file)
            save_file.unlink()
            feature_chunks.append(batch_features)
        
        save_dir.rmdir()

        features = np.concatenate(feature_chunks, axis=0)
        return features
````

## 4.2 Model stage

In this project, multiple machine learning models are implemented to explore different approaches for deception detection in narrated stories. Each model is selected based on its unique strengths and suitability for the classification task. Below is a brief description of each implemented model:

````python
from abc import ABC, abstractmethod
import warnings
from sklearnex import patch_sklearn

patch_sklearn() # apply the monkey patch from Intel to speed up scikit-learn
warnings.filterwarnings('ignore', category=FutureWarning, module='sklearn') # ignore FutureWarnings from scikit-learn, to suppress deprecation warnings

class Model(ABC):
    """
    Abstract class for all models.
    """
    @staticmethod
    @abstractmethod
    def create(**kwargs) -> 'Model':
        """
        Create a new instance of the model.
        """
        pass

    @abstractmethod
    def fit(self, X, y, **kwargs):
        """
        Fit the model on the training data.
        
        Args:
            X: training data
            y: training labels
            kwargs: additional arguments
        """
        pass

    @abstractmethod
    def predict(self, X) -> np.ndarray:
        """
        Predict the labels of the test data.
        
        Args:
            X: test data

        Returns:
            np.ndarray: predicted labels
        """
        pass

    @abstractmethod
    def predict_proba(self, X) -> np.ndarray:
        """
        Predict the probabilities of the test data.

        Args:
            X: test data

        Returns:
            np.ndarray: predicted probabilities
        """
        pass

    @abstractmethod
    def score(self, X, y) -> float:
        """
        Evaluate the model on the test data.

        Args:
            X: test data
            y: test labels

        Returns:
            float: accuracy score
        """
        pass

    @abstractmethod
    def save(self, path):
        """
        Save the model to a file.
        
        Args:
            path: path to save the model
        """
        pass

    @staticmethod
    @abstractmethod
    def load(path) -> 'Model':
        """
        Load the model from a file.

        Args:
            path: path to load the model
        
        Returns:
            Model: loaded model
        """
        pass
````

### K-Nearest Neighbors (KNN)
K-Nearest Neighbors is a straightforward, instance-based learning algorithm that classifies a data point based on the majority label among its k closest neighbors in the feature space. KNN is chosen for its simplicity and effectiveness in capturing local patterns within the data without requiring an explicit training phase. It serves as a strong baseline model for comparison with more complex algorithms.

````python
from sklearn.neighbors import KNeighborsClassifier
import joblib

class KNNModel(Model):
    knn: KNeighborsClassifier

    def __init__(self, model):
        self.knn = model

    @staticmethod
    def create(**kwargs) -> 'KNNModel':
        k = kwargs.get('k', 5)
        return KNNModel(KNeighborsClassifier(n_neighbors=k))
    
    def fit(self, X, y, **kwargs):
        self.knn.fit(X, y)

    def predict(self, X):
        return self.knn.predict(X)
    
    def predict_proba(self, X):
        return self.knn.predict_proba(X)
    
    def score(self, X, y):
        return self.knn.score(X, y)
    
    def save(self, path):
        joblib.dump(self.knn, path)

    @staticmethod
    def load(path) -> 'KNNModel':
        return KNNModel(joblib.load(path))
````

### Logistic Regression
Logistic Regression is a linear model commonly used for binary classification tasks. It estimates the probability that a given input belongs to a particular class by applying a logistic function to a linear combination of input features. Logistic Regression is favored for its interpretability, efficiency, and ability to provide probabilistic outputs, making it suitable for understanding feature importance in deception detection.

````python
from sklearn.linear_model import LogisticRegression
import joblib

class LogisticRegressionModel(Model):
    lr: LogisticRegression

    def __init__(self, model):
        self.lr = model

    @staticmethod
    def create(**kwargs) -> 'LogisticRegressionModel':
        C = kwargs.get('C', 1.0)
        random_state = kwargs.get('random_state', 42)
        max_iter = kwargs.get('max_iter', 1000)
        tol = kwargs.get('tol', 1e-4)
        return LogisticRegressionModel(
            LogisticRegression(
                penalty='l2', 
                C=C, 
                random_state=random_state,
                max_iter=max_iter, 
                tol=tol
            )
        )
    
    def fit(self, X, y, **kwargs):
        self.lr.fit(X, y)

    def predict(self, X):
        return self.lr.predict(X)
    
    def predict_proba(self, X):
        return self.lr.predict_proba(X)
    
    def score(self, X, y):
        return self.lr.score(X, y)
    
    def save(self, path):
        joblib.dump(self.lr, path)

    @staticmethod
    def load(path) -> 'LogisticRegressionModel':
        return LogisticRegressionModel(joblib.load(path))
````

### Random Forest
Random Forest is an ensemble learning method that constructs multiple decision trees during training and outputs the mode of their predictions. By aggregating the results of numerous trees, Random Forest reduces the risk of overfitting and enhances generalization performance. Its ability to handle high-dimensional data and provide insights into feature importance makes it a powerful tool for classification tasks like deception detection.

````python
from sklearn.ensemble import RandomForestClassifier
import joblib

class RandomForestModel(Model):
    rf: RandomForestClassifier

    def __init__(self, model):
        self.rf = model

    @staticmethod
    def create(**kwargs) -> 'RandomForestModel':
        n_estimators = kwargs.get('n_estimators', 100)
        random_state = kwargs.get('random_state', 42)
        return RandomForestModel(RandomForestClassifier(n_estimators=n_estimators, random_state=random_state))
    
    def fit(self, X, y, **kwargs):
        self.rf.fit(X, y)

    def predict(self, X):
        return self.rf.predict(X)
    
    def predict_proba(self, X):
        return self.rf.predict_proba(X)
    
    def score(self, X, y):
        return self.rf.score(X, y)
    
    def save(self, path):
        joblib.dump(self.rf, path)

    @staticmethod
    def load(path) -> 'RandomForestModel':
        return RandomForestModel(joblib.load(path))
````

### Recurrent Neural Network (RNN)
Recurrent Neural Networks, specifically those utilizing Long Short-Term Memory (LSTM) layers, are designed to capture temporal dependencies and sequential patterns within data. In the context of audio analysis, RNNs excel at modeling the temporal dynamics of spoken narratives, allowing the model to learn intricate patterns that may indicate deception. The ability of RNNs to handle varying input lengths and maintain context over time makes them well-suited for processing audio recordings in deception detection.

````python
import os
os.environ["KERAS_BACKEND"] = "torch"
from keras import Input, Sequential, layers, optimizers, Model as KerasModel

class RNNModel(Model):
    model: KerasModel

    def __init__(self, model):
        self.model = model

    @staticmethod
    def create(**kwargs) -> 'RNNModel':
        X_train_shape = kwargs.get('X_train_shape')
        y_train_len = kwargs.get('y_train_len')

        # Define the RNN model using Input layer
        model = Sequential()

        # Input layer
        model.add(Input(shape=X_train_shape))

        # Convolutional layer block 1
        model.add(layers.Conv1D(filters=64, kernel_size=3, padding='same'))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.MaxPooling1D(pool_size=2))
        model.add(layers.Dropout(0.3))

        # Convolutional layer block 2
        model.add(layers.Conv1D(filters=128, kernel_size=3, padding='same'))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.MaxPooling1D(pool_size=2))
        model.add(layers.Dropout(0.3))

        # Convolutional layer block 3
        model.add(layers.Conv1D(filters=256, kernel_size=3, padding='same'))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('relu'))
        model.add(layers.MaxPooling1D(pool_size=2))
        model.add(layers.Dropout(0.3))

        # Recurrent layer
        model.add(layers.Bidirectional(layers.LSTM(128, return_sequences=False, dropout=0.3, recurrent_dropout=0.3)))

        # Fully connected layer
        model.add(layers.Dense(128, activation='relu'))
        model.add(layers.Dropout(0.5))

        # Output layer
        model.add(layers.Dense(y_train_len, activation='softmax'))

        # Compile the model using sparse categorical crossentropy
        model.compile(optimizer=optimizers.Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        return RNNModel(model)

    
    def fit(self, X, y, **kwargs):
        epochs = kwargs.get('epochs', 100)
        batch_size = kwargs.get('batch_size', 128)
        validation_data = kwargs.get('validation_data', None)
        self.model.fit(X, y, epochs=epochs, batch_size=batch_size, validation_data=validation_data)

    def predict(self, X):
        return self.predict_proba(X).argmax(axis=1)
    
    def predict_proba(self, X):
        return self.model.predict(X)
    
    def score(self, X, y):
        return self.model.evaluate(X, y)
    
    def save(self, path):
        self.model.save(path)

    @staticmethod
    def load(path) -> 'RNNModel':
        from keras import models
        model = models.load_model(path)
        return RNNModel(model)
````

### Support Vector Machine (SVM)
Support Vector Machine is a robust classification algorithm that finds the optimal hyperplane separating different classes in the feature space. SVM is particularly effective in high-dimensional spaces and is capable of modeling complex, non-linear decision boundaries through the use of kernel functions. Its strong theoretical foundation and versatility make SVM a valuable model for distinguishing between true and false narratives.

````python
from sklearn.svm import SVC
import joblib

class SVMModel(Model):
    svm: SVC

    def __init__(self, model):
        self.svm = model

    @staticmethod
    def create(**kwargs):
        C = kwargs.get('C', 1.0)
        kernel = kwargs.get('kernel', 'rbf')
        return SVMModel(SVC(C=C, kernel=kernel,probability=True))
    
    def fit(self, X, y, **kwargs):
        self.svm.fit(X, y)

    def predict(self, X):
        return self.svm.predict(X)
    
    def predict_proba(self, X):
        return self.svm.predict_proba(X)
    
    def score(self, X, y):
        return self.svm.score(X, y)
    
    def save(self, path):
        joblib.dump(self.svm, path)

    @staticmethod
    def load(path) -> 'SVMModel':
        return SVMModel(joblib.load(path))
````

## 4.3 Ensemble stage

An ensemble approach combines the predictions of multiple individual models to enhance overall classification performance. By leveraging the strengths of diverse algorithms, the ensemble model aims to achieve higher accuracy and robustness compared to any single model. In this project, models such as KNN, Logistic Regression, Random Forest, and SVM are integrated using a meta-model (Logistic Regression) to aggregate their predictions effectively.

````python
from pathlib import Path
import numpy as np
from sklearn.linear_model import LogisticRegression

default_base_dir = Path.cwd() / 'weights'

class Ensembler:
    active_models: list[Model] = [] # list of active models
    meta_model: LogisticRegression # meta model, using Logistic Regression to ensemble the models
    base_dir: Path # base directory to save the meta model

    def __init__(self, meta_model: LogisticRegression, models: list[Model], base_dir=default_base_dir):
        """
        Initialize the Ensembler with a meta model and a list of active models.
        
        Args:
            meta_model: meta model to ensemble the active models
            models: list of active models
            base_dir: base directory to save the meta model
        """
        self.meta_model = meta_model
        self.base_dir = base_dir
        self.active_models = models
    
    @staticmethod
    def create(models: list[Model], base_dir=default_base_dir) -> 'Ensembler':
        """
        Create a new instance of the Ensembler.

        Args:
            models: list of active models
            base_dir: base directory to save the meta model

        Returns:
            Ensembler: new instance of the Ensembler
        """
        meta_model = LogisticRegression(max_iter=1000)
        return Ensembler(meta_model, models, base_dir)
    
    def fit(self, X: list[np.ndarray], labels):
        """
        Fit the Ensembler on the training data.
        
        Args:
             X: list of training data
            labels: training labels
        """
        X_meta_train = []

        for idx, model in enumerate(self.active_models):
            X_model = X[idx]
            x_meta = model.predict_proba(X_model)
            X_meta_train.append(x_meta)
        
        X_meta_train = np.hstack(X_meta_train)
        self.meta_model.fit(X_meta_train, labels)

    @staticmethod
    def load(meta_model_path: Path, models: list[Model]) -> 'Ensembler':
        """
        Load the Ensembler from a file.

        Args:
            meta_model_path: path to the meta model
            models: list of active models

        Returns:
            Ensembler: loaded Ensembler
        """
        meta_model = joblib.load(meta_model_path)
        return Ensembler(meta_model, models)
    
    def save(self, path: Path):
        """
        Save the Ensembler to a file.

        Args:
            path: path to save the Ensembler
        """
        joblib.dump(self.meta_model, path)
    
    def predict(self, X: list[np.ndarray]) -> np.ndarray:
        """
        Predict the labels of the test data.

        Args:
            X: list of test data

        Returns:
            np.ndarray: predicted labels
        """
        X_meta_test = []
        for idx, model in enumerate(self.active_models):
            X_model = X[idx]
            x_meta = model.predict_proba(X_model)
            X_meta_test.append(x_meta)
        
        X_meta_test = np.hstack(X_meta_test)
        return self.meta_model.predict(X_meta_test)
    
    def predict_proba(self, X: list[np.ndarray]) -> float:
        """
        Predict the probabilities of the test data.

        Args:
            X: list of test data

        Returns:
            float: predicted probabilities
        """
        X_meta_test = []
        for idx, model in enumerate(self.active_models):
            X_model = X[idx]
            x_meta = model.predict_proba(X_model)
            X_meta_test.append(x_meta)
        
        X_meta_test = np.hstack(X_meta_test)
        return self.meta_model.predict_proba(X_meta_test)
````

### 5.1 Data Preprocessing
The foundation of this project is the MLEnd Deception Dataset, comprising 100 audio recordings of narrated stories, each labeled as True or False. To enhance the dataset's diversity and improve the robustness of the machine learning models, data augmentation techniques are employed using the provided expand_audio_data function. This function systematically augments each original audio sample by applying the following transformations:

* Time Stretching: Alters the speed of the audio without changing its pitch, allowing the model to generalize across different speaking rates.
* Pitch Shifting: Modifies the pitch of the audio, enabling the model to handle variations in vocal tones.
* Adding Noise: Introduces background noise to simulate real-world audio environments, enhancing the model's resilience to noisy inputs.
* Volume Adjustment: Changes the audio volume to mimic different recording conditions.
* Frequency Masking: Applies masking to specific frequency bands, promoting the model's ability to focus on essential spectral features.
* Each original audio file is augmented fivefold, resulting in a total of 600 audio samples. This expansion mitigates overfitting and provides the models with a more comprehensive set of examples to learn from.


````python
from pathlib import Path
import shutil
import pandas as pd
import librosa
import numpy as np
import soundfile as sf
from tqdm import tqdm

def expand_audio_data():
    # Set up paths
    root_path = Path(__file__).parent.parent / 'dataset'
    raw_path = root_path / 'raw'
    ori_path = root_path / 'expanded'

    # Define mappings
    lang_mapping = {
        'Chinese': 'ZH',
        'English': 'EN',
    }
    type_mapping = {
        'True Story': 'T',
        'Deceptive Story': 'F'
    }

    # Read CSV attributes
    attributes = pd.read_csv(raw_path / 'story_attributes.csv')

    # Clear and create the expanded directory
    shutil.rmtree(ori_path, ignore_errors=True)
    ori_path.mkdir(parents=True, exist_ok=True)

    # Data augmentation functions
    def time_stretch(y, rate=1.2):
        return librosa.effects.time_stretch(y, rate=rate)

    def pitch_shift(y, sr, n_steps=4):
        return librosa.effects.pitch_shift(y, sr=sr, n_steps=n_steps)

    def add_noise(y, noise_level=0.005):
        noise = np.random.randn(len(y))
        return y + noise_level * noise

    def change_volume(y, factor=1.2):
        return y * factor

    def frequency_masking(y, frequency_mask_param=10):
        S = librosa.stft(y)
        mask = np.ones_like(S)
        num_freqs = S.shape[0]
        freq_mask_start = np.random.randint(0, num_freqs - frequency_mask_param)
        mask[freq_mask_start:freq_mask_start + frequency_mask_param, :] = 0
        return librosa.istft(S * mask)

    # Process and augment audio
    def process_audio(file):
        y, sr = librosa.load(file, sr=16000)

        augmented_audios = []
        augmented_audios.append(time_stretch(y))  # Time stretch
        augmented_audios.append(pitch_shift(y, sr))  # Pitch shift
        augmented_audios.append(add_noise(y))  # Add noise
        augmented_audios.append(change_volume(y))  # Change volume
        augmented_audios.append(frequency_masking(y))  # Frequency masking

        data = [y] + augmented_audios
        for i in range(len(data)):
            data[i] = librosa.util.fix_length(data[i], size=sr*300)
        return data

    # Expand wav files
    cnt = 1
    new_file_format = '{num}_{lang}_{type}_{augmented}.wav'

    for _, row in tqdm(attributes.iterrows(), total=attributes.shape[0], desc="Processing audio files"):
        file = Path(raw_path / row['filename'])
        file_type = type_mapping[row['Story_type']]
        lang = lang_mapping[row['Language']]

        if not file.exists():
            continue

        # Process the original file and its augmentations
        augmented_audios = process_audio(file)

        for i, audio in enumerate(augmented_audios):
            # Construct a new filename
            augmented_file = new_file_format.format(
                num=file.stem,
                lang=lang,
                type=file_type,
                augmented=f"aug{i + 1}"
            )
            augmented_file_path = ori_path / augmented_file

            # Save the augmented file
            sf.write(augmented_file_path, audio, 16000)
            cnt += 1
````

### Ensuring Data Integrity and Consistency
Throughout the data loading and structuring phase, careful attention is paid to handling missing or corrupted files. Any incomplete or unusable audio samples are identified and excluded, maintaining the dataset's integrity. Additionally, label encoding transforms categorical attributes, such as language and story type, into numerical formats suitable for machine learning algorithms. This encoding facilitates seamless integration with various models, ensuring that the data is both accurate and compatible with the chosen analytical methods.

````python
from pathlib import Path
import numpy as np
import torch
import torchaudio
import librosa
import soundfile
from joblib import Memory
from dataclasses import dataclass

dataset_path = Path.cwd() / 'dataset' / 'expanded'

@dataclass
class Audio:
    audio: np.ndarray | torch.Tensor
    sr: int
    language: np.ushort
    is_true: bool

memory = Memory('/tmp', verbose=0)

def __load_data(backend='librosa', only_labels=False) -> tuple[list[Audio], list[bool]]:
    """
    Load the data from the dataset folder
    :return: List of tuples containing the librosa-loaded audio, sample rate, one-hot encoded language, and if it is a true story
    """
    dat = []
    label = []
    lang_mapping = {
        'ZH': 0,
        'EN': 1,
    }
    for path in dataset_path.iterdir():
        if only_labels:
            label.append(path.stem.split('_')[2] == 'T')
            continue
        if backend == 'librosa':
            audio, sr = librosa.load(path, sr=None)
        elif backend == 'torchaudio':
            audio, sr = torchaudio.load(path)
        elif backend == 'soundfile':
            audio, sr = soundfile.read(path)
        else:
            raise ValueError(f'Invalid backend: {backend}')
        idx, language, story_type, _ = path.stem.split('_', 3)
        language = np.ushort(lang_mapping[language])
        story_type = story_type == 'T'
        dat.append(Audio(audio, sr, language, story_type))
        label.append(story_type)
    return dat, label

load_data = memory.cache(__load_data)
````

### Evaluation
The evaluation class is designed to comprehensively assess the performance of machine learning models in distinguishing between true and false narrated stories. It systematically calculates essential evaluation metrics such as accuracy, sensitivity (recall), specificity, precision, and the F1 score, which collectively provide a detailed understanding of the model's classification capabilities. Additionally, the class generates visual representations like confusion matrices and Receiver Operating Characteristic (ROC) curves with their corresponding Area Under the Curve (AUC) values. These visual tools help in identifying the strengths and weaknesses of the model by illustrating how well it differentiates between the classes and how it performs across various threshold settings. By integrating both quantitative metrics and graphical analyses, the Evaluator class facilitates a holistic evaluation process, enabling researchers to make informed decisions about model selection, tuning, and potential areas for improvement. This structured approach ensures that the models are not only accurate but also reliable and robust in real-world applications of deception detection in audio narratives.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc

class Evaluator:
    model: Model
    X_test: np.ndarray
    y_test: np.ndarray

    def __init__(self, model: Model, X_test: np.ndarray, y_test: np.ndarray):
        self.model = model
        self.X_test = X_test
        self.y_test = y_test

    def print_metrics(self):
        self.print_accuracy()
        self.print_class_sensitive_rates()
        self.plot_confusion_matrix()
        self.plot_roc_auc()

    @property
    def accuracy(self):
        y_pred = self.model.predict(self.X_test)
        return accuracy_score(self.y_test, y_pred)
    
    def print_accuracy(self):
        print(f'Accuracy: {self.accuracy:.2f}')
    
    @property
    def class_sensitive_rates(self):
        y_pred = self.model.predict(self.X_test)
        cm = confusion_matrix(self.y_test, y_pred)
        tp = cm[0, 0]
        tn = cm[1, 1]
        fp = cm[0, 1]
        fn = cm[1, 0]
        sensitivity = tp / (tp + fn)
        specificity = tn / (tn + fp)
        precision = tp / (tp + fp)
        f1_score = 2 * (precision * sensitivity) / (precision + sensitivity)
        return sensitivity, specificity, precision, f1_score
    
    def print_class_sensitive_rates(self):
        sensitivity, specificity, precision, f1_score = self.class_sensitive_rates
        print(f'Sensitivity: {sensitivity:.2f}')
        print(f'Specificity: {specificity:.2f}')
        print(f'Precision: {precision:.2f}')
        print(f'F1 Score: {f1_score:.2f}')
    
    @property
    def confusion_matrix(self):
        y_pred = self.model.predict(self.X_test)
        return confusion_matrix(self.y_test, y_pred)

    def plot_confusion_matrix(self):
        class_labels = ['Positive', 'Negative']
        cm = pd.DataFrame(self.confusion_matrix, index=class_labels, columns=class_labels)
        plt.figure()
        sns.heatmap(cm, annot=True, fmt='d')
        plt.xlabel('True')
        plt.ylabel('Predicted')
        plt.title('Confusion Matrix')
        plt.show()
    
    @property
    def roc_auc(self) -> tuple[np.ndarray, np.ndarray, float]:
        y_pred = self.model.predict_proba(self.X_test)[:, 1]
        fpr, tpr, _ = roc_curve(self.y_test, y_pred)
        roc_auc = auc(fpr, tpr)
        return fpr, tpr, roc_auc
    
    def plot_roc_auc(self):
        import matplotlib.pyplot as plt
        fpr, tpr, roc_auc = self.roc_auc
        plt.figure()
        plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('1-specificity')
        plt.ylabel('sensitivity')
        plt.title('Receiver Operating Characteristic (ROC) Curve')
        plt.legend(loc='lower right')
        plt.show()
````

This division ensures enough data for training while retaining sufficient samples for reliable testing. The training set allows models to learn patterns distinguishing true from deceptive narratives, while the testing set provides an unbiased evaluation of the models' ability to generalize to unseen data. To maintain dataset integrity, random shuffling is applied before splitting, ensuring an independent and identically distributed (IID) sample. ### Ensemble Model
This part is to ensemble the pre-trained models using Logistic Regression. The pre-trained models are SVM, Random Forest, and KNN. The ensemble model is Logistic Regression.
The training of the ensemble model is performed by invoking each pre-trained model, obtaining the predicted probabilities, and then training the ensemble model using the predicted probabilities.

````python
ensemble_model_path = weights_path / 'ensemble_model.joblib'
use_cached_ensembler = True

# rnn_model is disable because of it's bad performance
if use_cached_ensembler:
    ensembler = Ensembler.load(ensemble_model_path, [knn_model, lr_model, rf_model, svm_model])
else:
    ensembler = Ensembler.create([knn_model, lr_model, rf_model, svm_model])
    ensembler.fit([flattened_X_train, flattened_X_train, flattened_X_train, flattened_X_train, plain_X_train], y_train)
    ensembler.save(ensemble_model_path)

ensemble_evaluator = Evaluator(ensembler, [flattened_X_test, flattened_X_test, flattened_X_test, flattened_X_test, plain_X_test], y_test)
ensemble_evaluator.print_metrics()
````
